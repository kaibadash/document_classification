```python
import glob
import os
import random
import pickle
import MeCab
import pandas as pd
from scdv import SparseCompositeDocumentVectors, build_word2vec
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from document import Document

tagger = MeCab.Tagger('-Owakati')
model = pickle.load(open("model/model.pkl", "rb"))
scdv = SparseCompositeDocumentVectors(
    model,
    2,
    100,
    "model/gmm_cluster.pkl",
    "model/gmm_prob_cluster.pkl"
)

files = glob.glob("tmp/text/**/*.txt")
tagger = MeCab.Tagger('-Owakati')
documents = []
df = pd.DataFrame()
for file_path in files:
    # 少ないデータで試す
    if random.randrange(100) != 0:
        print("skip: " + file_path)
        continue
    print("load: " + file_path)

    with open(file_path, 'r') as file:
        text = file.read()
    lines = text.split("\n")
    lines.pop(0)
    lines.pop(0)
    category = file_path.split("/")[-2]
    sentences = []
    for line in lines:
        if line.rstrip() == "":
            continue
        sentences.append(tagger.parse(line).strip().split())
    scdv.get_probability_word_vectors(sentences)
    print([category, scdv.make_gwbowv(sentences)])
    print(Document(category, scdv.make_gwbowv(sentences)).toArray())
    df.loc[:] = Document(category, scdv.make_gwbowv(sentences)).toArray()
```
